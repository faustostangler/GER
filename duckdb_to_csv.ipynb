```python
import duckdb
import pandas as pd

# 1. Configuração de Caminhos
parquet_file = "gercon_consolidado.parquet"
output_csv = "gercon_duckdb.csv"

print(f"--- 📊 Iniciando Extração de Amostra para Inspeção ---")

# 2. Conexão DuckDB e Extração via SQL
# Utilizamos ROW_NUMBER() para garantir exatamente 500 registros por "Origem da Lista"
con = duckdb.connect(database=':memory:')

query = f"""
SELECT * EXCLUDE (row_num)
FROM (
    SELECT *, 
           ROW_NUMBER() OVER (
               PARTITION BY "Origem da Lista" 
               ORDER BY "Data Solicitação" DESC
           ) as row_num
    FROM read_parquet('{parquet_file}')
) 
WHERE row_num <= 500
"""

try:
    # 3. Execução e Conversão para DataFrame
    df_amostra = con.execute(query).df()
    
    # 4. Exportação para CSV (Formatado para Excel no Brasil)
    # Usamos sep=';' e encoding='utf-8-sig' para que o Excel abra corretamente acentos e colunas
    df_amostra.to_csv(output_csv, sep=';', index=False, encoding='utf-8-sig')
    
    print(f"✅ Sucesso! Arquivo '{output_csv}' gerado.")
    print(f"📈 Total de registros extraídos: {len(df_amostra)}")
    
    # Exibir contagem por origem para conferência
    print("\nResumo da Amostra por Origem:")
    print(df_amostra["Origem da Lista"].value_counts())

except Exception as e:
    print(f"❌ Erro ao processar o Parquet: {e}")
```

```python
import pandas as pd

def fast_gercon_parser(data):
    """
    Parser dinâmico e simples. 
    Transforma JSONs complexos do GERCON em um DataFrame plano (flat).
    """
    # 1. Garantir que os dados estão em formato de lista de dicionários
    if isinstance(data, str):
        data = json.loads(data)
    if isinstance(data, dict):
        data = [data]
        
    # 2. Flattening Automático: Transforma {"usuarioSUS": {"nome": "X"}} em "usuarioSUS_nome"
    # O separador '_' mantém a compatibilidade com sua Linguagem Ubíqua
    df = pd.json_normalize(data, sep='_')
    
    # 3. Conversão de Timestamps (O GERCON usa milissegundos)
    # Identifica colunas de data/versão e converte para formato legível
    date_cols = [c for c in df.columns if any(x in c.lower() for x in ['data', 'versao'])]
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], unit='ms', errors='coerce')
        
    return df

df = fast_gercon_parser(registros)
df.T.to_csv("json.csv")
```

```python
import duckdb
import pandas as pd

# Conectar ao motor em memória
con = duckdb.connect()
parquet_file = "gercon_consolidado.parquet"

print("🔍 Iniciando Validação de Integridade da Camada Gold...\n")

try:
    # --- 1. Teste de Volumetria ---
    total = con.execute(f"SELECT count(*) FROM read_parquet('{parquet_file}')").fetchone()[0]
    print(f"📊 Volumetria Total: {total} registos.")
    if total == 0:
        print("❌ ERRO: O ficheiro Parquet está vazio!")

    # --- 2. Teste de Clean Break (Zero Legado) ---
    schema_df = con.execute(f"DESCRIBE SELECT * FROM read_parquet('{parquet_file}')").df()
    colunas = schema_df['column_name'].tolist()
    
    colunas_legadas = ["Protocolo", "Risco Cor", "Município de Residência", "Médico Solicitante", "Situação Final"]
    legado_encontrado = [c for c in colunas_legadas if c in colunas]
    
    if legado_encontrado:
        print(f"❌ FALHA DE CLEAN BREAK: Colunas antigas ainda existem: {legado_encontrado}")
    else:
        print("✅ Clean Break Confirmado: Zero colunas legadas no Parquet (Linguagem Ubíqua aplicada).")

# --- 3. Auditoria do Motor SLA (Ping-Pong) ---
    print("\n⏱️ Auditoria do Motor de SLA (Médias Globais):")
    sla_stats = con.execute(f"""
        SELECT 
            ROUND(AVG(CAST(SLA_Lead_Time_Total_Dias AS DOUBLE)), 1) as media_lead_time_dias,
            MAX(CAST(SLA_Interacoes_Regulacao AS INTEGER)) as maximo_ping_pong,
            COUNT(*) FILTER (WHERE SLA_Desfecho_Atingido = true) as qtd_casos_encerrados
        FROM read_parquet('{parquet_file}')
        WHERE SLA_Lead_Time_Total_Dias IS NOT NULL
    """).df()
    print(sla_stats.to_string(index=False))

    # --- 4. Auditoria do Snapshot (Entidade) ---
    print("\n🏥 Verificação de Integridade Clínica (Top 3 Registos):")
    amostra = con.execute(f"""
        SELECT 
            numeroCMCE, 
            entidade_especialidade_descricao as especialidade, 
            entidade_classificacaoRisco_cor as risco
        FROM read_parquet('{parquet_file}')
        WHERE entidade_especialidade_descricao IS NOT NULL AND entidade_especialidade_descricao != ''
        LIMIT 3
    """).df()
    print(amostra.to_string(index=False))

except Exception as e:
    print(f"❌ Falha Crítica ao ler o Parquet: {e}")
```

```python
import os
import pandas as pd
import duckdb

# 1. Conexão com DuckDB e Carregamento (SRE pattern: DuckDB is the engine)
con = duckdb.connect()
arquivo_parquet = "gercon_consolidado.parquet"
df = con.execute(f"SELECT * FROM '{arquivo_parquet}'").df()

# 2. Pasta de Auditoria Volumétrica
output_dir = "inspecao_volumetria_unicos"
os.makedirs(output_dir, exist_ok=True)

print(f"--- 📊 Iniciando Auditoria Volumétrica ({len(df.columns)} colunas) ---")

for col in df.columns:
    print(f"Processando: {col}")
    
    # Limpa o nome do ficheiro
    clean_filename = col.replace(" ", "_").replace("/", "-").replace("\\", "-").replace("?", "")
    path = os.path.join(output_dir, f"{clean_filename}.csv")
    
    # 🔍 MÁGICA: Transforma a série de únicos em um DataFrame de Frequência
    # dropna=False permite auditar quantos campos estão vazios no Data Lake
    df_volumetria = df[col].value_counts(dropna=False).reset_index()
    df_volumetria.columns = [col, 'Quantidade']
    
    # 3. Exporta para CSV (ordenado pelo item mais frequente primeiro)
    df_volumetria.to_csv(path, index=False, sep=';', encoding='utf-8-sig')

print(f"✅ Sucesso! {len(df.columns)} ficheiros com volumetria gerados em '{output_dir}/'.")
```
