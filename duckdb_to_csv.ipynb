```python
import duckdb
import pandas as pd

# 1. Configuração de Caminhos
parquet_file = "gercon_consolidado.parquet"
output_csv = "gercon_duckdb.csv"

print("--- 📊 Iniciando Extração de Amostra para Inspeção ---")

# 2. Conexão DuckDB e Extração via SQL
# Utilizamos ROW_NUMBER() para garantir exatamente 500 registros por "Origem da Lista"
con = duckdb.connect(database=":memory:")

query = f"""
SELECT * EXCLUDE (row_num)
FROM (
    SELECT *, 
           ROW_NUMBER() OVER (
               PARTITION BY "Origem da Lista" 
               ORDER BY "Data Solicitação" DESC
           ) as row_num
    FROM read_parquet('{parquet_file}')
) 
WHERE row_num <= 500
"""

try:
    # 3. Execução e Conversão para DataFrame
    df_amostra = con.execute(query).df()

    # 4. Exportação para CSV (Formatado para Excel no Brasil)
    # Usamos sep=';' e encoding='utf-8-sig' para que o Excel abra corretamente acentos e colunas
    df_amostra.to_csv(output_csv, sep=";", index=False, encoding="utf-8-sig")

    print(f"✅ Sucesso! Arquivo '{output_csv}' gerado.")
    print(f"📈 Total de registros extraídos: {len(df_amostra)}")

    # Exibir contagem por origem para conferência
    print("\nResumo da Amostra por Origem:")
    print(df_amostra["Origem da Lista"].value_counts())

except Exception as e:
    print(f"❌ Erro ao processar o Parquet: {e}")
```

```python
def fast_gercon_parser(data):
    """
    Parser dinâmico e simples.
    Transforma JSONs complexos do GERCON em um DataFrame plano (flat).
    """
    # 1. Garantir que os dados estão em formato de lista de dicionários
    if isinstance(data, str):
        data = json.loads(data)
    if isinstance(data, dict):
        data = [data]

    # 2. Flattening Automático: Transforma {"usuarioSUS": {"nome": "X"}} em "usuarioSUS_nome"
    # O separador '_' mantém a compatibilidade com sua Linguagem Ubíqua
    df = pd.json_normalize(data, sep="_")

    # 3. Conversão de Timestamps (O GERCON usa milissegundos)
    # Identifica colunas de data/versão e converte para formato legível
    date_cols = [
        c for c in df.columns if any(x in c.lower() for x in ["data", "versao"])
    ]
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], unit="ms", errors="coerce")

    return df


df = fast_gercon_parser(registros)
df.T.to_csv("json.csv")
```

```python
import pandas as pd
import numpy as np
import json
from pandas.api.types import (
    is_numeric_dtype,
    is_datetime64_any_dtype,
    is_bool_dtype,
    is_string_dtype,
    is_object_dtype,
)

OUTPUT_CSV = "perfil_colunas.csv"
OUTPUT_XLSX = "perfil_colunas.xlsx"
TOP_N_VALUES = 5

SENSITIVE_HINTS = [
    "cpf",
    "nome",
    "mae",
    "cartao",
    "sus",
    "logradouro",
    "bairro",
    "cep",
    "complemento",
    "numero",
    "nascimento",
    "quadro_clinico",
    "evolucao",
    "historico",
]

TEXT_NULL_LIKE = {
    "",
    " ",
    "nan",
    "none",
    "null",
    "na",
    "n/a",
    "não informado",
    "nao informado",
}


def safe_sample(series: pd.Series, n: int = 3):
    return series.dropna().astype(str).head(n).tolist()


def classify_cardinality(n_unique: int, total_rows: int, unique_ratio: float) -> str:
    if total_rows == 0:
        return "indefinida"
    if n_unique == 0:
        return "sem valores"
    if n_unique == 1:
        return "constante"
    if unique_ratio >= 0.95:
        return "quase unica"
    if unique_ratio >= 0.50:
        return "alta"
    if unique_ratio >= 0.05:
        return "media"
    return "baixa"


def detect_sensitive(col_name: str) -> bool:
    name = col_name.lower()
    return any(token in name for token in SENSITIVE_HINTS)


def get_top_values(series: pd.Series, top_n: int = 5):
    vc = series.dropna().astype(str).value_counts(dropna=True).head(top_n)
    if vc.empty:
        return []
    return [{"valor": idx, "freq": int(val)} for idx, val in vc.items()]


def text_empty_stats(series: pd.Series):
    s = series.dropna().astype(str).str.strip().str.lower()
    if s.empty:
        return 0, 0.0
    empty_like = s.isin(TEXT_NULL_LIKE).sum()
    pct = empty_like / len(series) * 100 if len(series) else 0.0
    return int(empty_like), pct


def looks_like_datetime(
    series: pd.Series, sample_size: int = 200, threshold: float = 0.8
):
    """
    Testa só uma amostra de strings não nulas para evitar warnings e custo alto.
    """
    s = series.dropna()
    if s.empty:
        return False, 0.0

    sample = s.astype(str).head(sample_size)

    # Heurística barata antes do parse
    # Só tenta parse se houver sinais de data em boa parte dos valores
    has_date_pattern = sample.str.contains(
        r"\d{1,4}[-/]\d{1,2}[-/]\d{1,4}|\d{1,2}:\d{2}|T\d{2}:\d{2}",
        regex=True,
        na=False,
    )
    if has_date_pattern.mean() < 0.4:
        return False, 0.0

    parsed = pd.to_datetime(sample, errors="coerce")
    ratio = parsed.notna().mean()
    return ratio >= threshold, float(ratio)


def infer_analytical_type(
    col_name: str, series: pd.Series, n_unique: int, unique_ratio: float
) -> str:
    name = col_name.lower()

    if is_datetime64_any_dtype(series):
        return "data_hora"

    if is_bool_dtype(series):
        return "booleana"

    if is_numeric_dtype(series) and not is_bool_dtype(series):
        if any(token in name for token in ["id", "cpf", "cartao", "cep", "cmce"]):
            return "identificador_potencial"
        if n_unique <= 2:
            return "booleana_codificada"
        if pd.api.types.is_integer_dtype(series.dropna()):
            return "numerica_discreta"
        return "numerica_continua"

    if any(
        token in name
        for token in ["cpf", "cartao", "numero", "id", "cmce", "cns", "cep"]
    ):
        return "identificador_potencial"

    if n_unique <= 2:
        return "booleana_textual"

    is_dt_like, _ = looks_like_datetime(series)
    if is_dt_like:
        return "data_hora_texto"

    avg_len = (
        series.dropna().astype(str).str.len().mean()
        if series.dropna().shape[0] > 0
        else 0
    )

    if unique_ratio >= 0.90:
        if avg_len > 30:
            return "texto_livre"
        return "alta_cardinalidade"

    if avg_len > 40:
        return "texto_livre"

    return "categorica"


def numeric_stats(series: pd.Series):
    """
    Estatísticas para colunas numéricas reais, excluindo boolean.
    """
    if is_bool_dtype(series):
        return {
            "min": np.nan,
            "p25": np.nan,
            "median": np.nan,
            "mean": np.nan,
            "p75": np.nan,
            "max": np.nan,
            "std": np.nan,
            "zeros": np.nan,
            "pct_zeros": np.nan,
            "negatives": np.nan,
            "pct_negatives": np.nan,
            "iqr": np.nan,
            "outliers_iqr": np.nan,
            "pct_outliers_iqr": np.nan,
        }

    s = pd.to_numeric(series, errors="coerce").dropna()

    if s.empty:
        return {
            "min": np.nan,
            "p25": np.nan,
            "median": np.nan,
            "mean": np.nan,
            "p75": np.nan,
            "max": np.nan,
            "std": np.nan,
            "zeros": np.nan,
            "pct_zeros": np.nan,
            "negatives": np.nan,
            "pct_negatives": np.nan,
            "iqr": np.nan,
            "outliers_iqr": np.nan,
            "pct_outliers_iqr": np.nan,
        }

    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = ((s < lower) | (s > upper)).sum()

    zeros = (s == 0).sum()
    negatives = (s < 0).sum()

    return {
        "min": s.min(),
        "p25": q1,
        "median": s.median(),
        "mean": s.mean(),
        "p75": q3,
        "max": s.max(),
        "std": s.std(),
        "zeros": int(zeros),
        "pct_zeros": zeros / len(series) * 100 if len(series) else np.nan,
        "negatives": int(negatives),
        "pct_negatives": negatives / len(series) * 100 if len(series) else np.nan,
        "iqr": iqr,
        "outliers_iqr": int(outliers),
        "pct_outliers_iqr": outliers / len(series) * 100 if len(series) else np.nan,
    }


def datetime_stats(series: pd.Series):
    s = pd.to_datetime(series, errors="coerce").dropna()
    if s.empty:
        return {
            "data_min": pd.NaT,
            "data_max": pd.NaT,
            "amplitude_dias": np.nan,
            "datas_futuras": np.nan,
            "pct_datas_futuras": np.nan,
            "datas_antes_2000": np.nan,
            "pct_datas_antes_2000": np.nan,
        }

    now = pd.Timestamp.now()
    future_count = (s > now).sum()
    very_old_count = (s < pd.Timestamp("2000-01-01")).sum()

    return {
        "data_min": s.min(),
        "data_max": s.max(),
        "amplitude_dias": (s.max() - s.min()).days,
        "datas_futuras": int(future_count),
        "pct_datas_futuras": future_count / len(series) * 100
        if len(series)
        else np.nan,
        "datas_antes_2000": int(very_old_count),
        "pct_datas_antes_2000": very_old_count / len(series) * 100
        if len(series)
        else np.nan,
    }


def boolean_stats(series: pd.Series):
    s = series.dropna()
    if s.empty:
        return {
            "bool_true": np.nan,
            "bool_false": np.nan,
            "bool_pct_true": np.nan,
            "bool_pct_false": np.nan,
        }

    # funciona para bool puro; se vier 0/1 como int, isso não entra aqui
    true_count = (s).sum()
    false_count = (not s).sum()

    return {
        "bool_true": int(true_count),
        "bool_false": int(false_count),
        "bool_pct_true": true_count / len(series) * 100 if len(series) else np.nan,
        "bool_pct_false": false_count / len(series) * 100 if len(series) else np.nan,
    }


def text_stats(series: pd.Series):
    s = series.dropna().astype(str)
    if s.empty:
        return {
            "len_min": np.nan,
            "len_mean": np.nan,
            "len_median": np.nan,
            "len_max": np.nan,
        }

    lengths = s.str.len()
    return {
        "len_min": lengths.min(),
        "len_mean": lengths.mean(),
        "len_median": lengths.median(),
        "len_max": lengths.max(),
    }


def build_observations(
    col_name: str,
    total_rows: int,
    n_missing: int,
    n_unique: int,
    unique_ratio: float,
    inferred_type: str,
    top1_freq: int | None,
    parsed_datetime_ratio: float | None = None,
):
    obs = []

    if n_missing == total_rows:
        obs.append("100% nula")

    if n_missing > 0 and total_rows > 0:
        pct_missing = n_missing / total_rows
        if pct_missing >= 0.90:
            obs.append("muitos nulos (>90%)")
        elif pct_missing >= 0.50:
            obs.append("muitos nulos (>50%)")

    if n_unique == 1 and total_rows > 0:
        obs.append("coluna constante")

    if unique_ratio >= 0.95 and total_rows > 0:
        obs.append("alta unicidade")

    if inferred_type == "identificador_potencial":
        obs.append("possivel identificador")

    if inferred_type == "texto_livre":
        obs.append("possivel texto livre")

    if (
        inferred_type in {"categorica", "booleana_textual", "booleana"}
        and top1_freq is not None
        and total_rows > 0
    ):
        dominance = top1_freq / total_rows
        if dominance >= 0.95:
            obs.append("categoria dominante >95%")
        elif dominance >= 0.80:
            obs.append("categoria dominante >80%")

    if detect_sensitive(col_name):
        obs.append("campo sensivel")

    if parsed_datetime_ratio is not None and 0.0 < parsed_datetime_ratio < 0.8:
        obs.append("texto com parte convertivel para data")

    return "; ".join(obs)


results = []
total_rows = len(df)

for col in df.columns:
    s = df[col]
    dtype_pandas = str(s.dtype)

    n_missing = int(s.isna().sum())
    pct_missing = (n_missing / total_rows * 100) if total_rows else np.nan

    n_unique = int(s.nunique(dropna=True))
    pct_unique = (n_unique / total_rows * 100) if total_rows else np.nan
    unique_ratio = (n_unique / total_rows) if total_rows else np.nan

    samples = safe_sample(s, n=3)

    empty_like_count = np.nan
    empty_like_pct = np.nan
    if is_string_dtype(s) or is_object_dtype(s):
        empty_like_count, empty_like_pct = text_empty_stats(s)

    inferred_type = infer_analytical_type(col, s, n_unique, unique_ratio)
    cardinality = classify_cardinality(n_unique, total_rows, unique_ratio)

    mode_value = None
    mode_freq = None
    top_values = get_top_values(s, top_n=TOP_N_VALUES)
    if top_values:
        mode_value = top_values[0]["valor"]
        mode_freq = top_values[0]["freq"]

    num_stats = {
        "min": np.nan,
        "p25": np.nan,
        "median": np.nan,
        "mean": np.nan,
        "p75": np.nan,
        "max": np.nan,
        "std": np.nan,
        "zeros": np.nan,
        "pct_zeros": np.nan,
        "negatives": np.nan,
        "pct_negatives": np.nan,
        "iqr": np.nan,
        "outliers_iqr": np.nan,
        "pct_outliers_iqr": np.nan,
    }

    dt_stats = {
        "data_min": pd.NaT,
        "data_max": pd.NaT,
        "amplitude_dias": np.nan,
        "datas_futuras": np.nan,
        "pct_datas_futuras": np.nan,
        "datas_antes_2000": np.nan,
        "pct_datas_antes_2000": np.nan,
    }

    bool_stats = {
        "bool_true": np.nan,
        "bool_false": np.nan,
        "bool_pct_true": np.nan,
        "bool_pct_false": np.nan,
    }

    txt_stats = {
        "len_min": np.nan,
        "len_mean": np.nan,
        "len_median": np.nan,
        "len_max": np.nan,
    }

    parsed_datetime_ratio = None

    if is_bool_dtype(s):
        bool_stats = boolean_stats(s)

    elif is_datetime64_any_dtype(s):
        dt_stats = datetime_stats(s)

    elif is_numeric_dtype(s):
        num_stats = numeric_stats(s)

    else:
        is_dt_like, parsed_datetime_ratio = looks_like_datetime(s)
        if is_dt_like:
            dt_stats = datetime_stats(s)

        txt_stats = text_stats(s)

    observations = build_observations(
        col_name=col,
        total_rows=total_rows,
        n_missing=n_missing,
        n_unique=n_unique,
        unique_ratio=unique_ratio,
        inferred_type=inferred_type,
        top1_freq=mode_freq,
        parsed_datetime_ratio=parsed_datetime_ratio,
    )

    results.append(
        {
            "coluna": col,
            "dtype_pandas": dtype_pandas,
            "tipo_inferido": inferred_type,
            "cardinalidade": cardinality,
            "total_linhas": total_rows,
            "nulos": n_missing,
            "pct_nulos": round(pct_missing, 4) if pd.notna(pct_missing) else np.nan,
            "vazios_textuais": empty_like_count,
            "pct_vazios_textuais": round(empty_like_pct, 4)
            if pd.notna(empty_like_pct)
            else np.nan,
            "n_unicos": n_unique,
            "pct_unicos": round(pct_unique, 4) if pd.notna(pct_unique) else np.nan,
            "valor_mais_frequente": mode_value,
            "freq_valor_mais_frequente": mode_freq,
            "amostra_valores": json.dumps(samples, ensure_ascii=False),
            "top_5_valores": json.dumps(top_values, ensure_ascii=False),
            "num_min": num_stats["min"],
            "num_p25": num_stats["p25"],
            "num_mediana": num_stats["median"],
            "num_media": num_stats["mean"],
            "num_p75": num_stats["p75"],
            "num_max": num_stats["max"],
            "num_std": num_stats["std"],
            "num_zeros": num_stats["zeros"],
            "num_pct_zeros": num_stats["pct_zeros"],
            "num_negativos": num_stats["negatives"],
            "num_pct_negativos": num_stats["pct_negatives"],
            "num_iqr": num_stats["iqr"],
            "num_outliers_iqr": num_stats["outliers_iqr"],
            "num_pct_outliers_iqr": num_stats["pct_outliers_iqr"],
            "data_min": dt_stats["data_min"],
            "data_max": dt_stats["data_max"],
            "data_amplitude_dias": dt_stats["amplitude_dias"],
            "data_futuras": dt_stats["datas_futuras"],
            "data_pct_futuras": dt_stats["pct_datas_futuras"],
            "data_antes_2000": dt_stats["datas_antes_2000"],
            "data_pct_antes_2000": dt_stats["pct_datas_antes_2000"],
            "bool_true": bool_stats["bool_true"],
            "bool_false": bool_stats["bool_false"],
            "bool_pct_true": bool_stats["bool_pct_true"],
            "bool_pct_false": bool_stats["bool_pct_false"],
            "txt_len_min": txt_stats["len_min"],
            "txt_len_media": txt_stats["len_mean"],
            "txt_len_mediana": txt_stats["len_median"],
            "txt_len_max": txt_stats["len_max"],
            "sensivel": detect_sensitive(col),
            "observacoes": observations,
        }
    )

perfil_df = pd.DataFrame(results)

perfil_df = perfil_df.sort_values(
    by=["tipo_inferido", "pct_nulos", "pct_unicos"], ascending=[True, False, False]
).reset_index(drop=True)

perfil_df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")
# perfil_df.to_excel(OUTPUT_XLSX, index=False)

print(f"Arquivo CSV salvo em: {OUTPUT_CSV}")
# print(f"Arquivo Excel salvo em: {OUTPUT_XLSX}")

display(perfil_df.head(20))
```
